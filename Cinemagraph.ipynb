```
import numpy as np
import tensorflow as tf

np.random.seed(3)
tf.random.set_seed(3)
```

```
import pandas as pd

df = pd.read_csv('/survey.csv', names= ['ID','placied','calm','serene','soothing','satisfying','tranquil','leisurely','lyrical','peaceful',
                                        'happy','glad','delighted','bright','cheerful','merry','pleased','vivacious','warmhearted','tense',
                                        'jittery','nervous','fearful','anxious','dignified','solemn','serious','sober','distressed','upset',
                                        'unhappy','miserable','grouchy','depressed','unpleasant','VisualExp','HearingExp','TasteExp','OlfactoryExp','TactileExp',
                                        'VisualExp2','HearingExp2','TasteExp2','OlfactoryExp2','TactileExp2','attitude1','attitude2','attitude3','attitude4','attitude5','image'])
print(df.head())
print(df.info())
```

```
dataset = df.values
X = dataset[:,0:50].astype(float)
Y_obj = dataset[:,50]
```

```
from sklearn.preprocessing import LabelEncoder

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
print(Y)

num = set(Y)
print(len(num)) # 사용된 이미지수
```

```
Y_encoded = tf.keras.utils.to_categorical(Y)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(16, input_dim=50, activation='relu'))
model.add(Dense(225, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```
epochs = 100
batch_size = 1

model.fit(X, Y_encoded, epochs=epochs, batch_size=batch_size)

print("\n Accuracy: %.4f"%(model.evaluate(X, Y_encoded)[1]))
```
