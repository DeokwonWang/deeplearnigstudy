```
import numpy as np
import matplotlib.pyplot as plt

# 과외 성적 포함
data=[[2, 0, 81],[4, 4, 93],[6, 2, 91],[8, 3, 97]]
x1=[i[0] for i in data]
x2=[i[1] for i in data]
y=[i[2] for i in data]

# 그래프
ax=plt.axes(projection='3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('score')
ax.dist=11
ax.scatter(x1,x2,y)
plt.show()
```

```
# 리스트의 데이터를 np.array 형식으로 변경
x1_data=np.array(x1)
x2_data=np.array(x2)
y_data=np.array(y)

# 기울기와 절편 초기화
a1 = 0
a2 = 0
b = 0

# 학습률
lr = 0.02

# 반복 회수 지정
epochs = 2001

# 경사하강법
for i in range(epochs):   # 에폭의 수 만큼 반복
  y_pred = a1 * x1_data + a2 * x2_data + b  # y 예측값을 구하는 식
  error= y_data - y_pred  # 오차를 구하는 식
  
  a1_diff = -(2/len(x1_data)) * sum(x1_data * error)
  a2_diff = -(2/len(x2_data)) * sum(x2_data * error)
  b_diff = -(2/len(x1_data)) * sum(error)

  a1 = a1 - lr * a1_diff # 학습률을 곱해 a1값 업데이트
  a2 = a2 - lr * a2_diff # 학습률을 곱해 a2값 업데이트
  b = b - lr * b_diff # 학습률을 곱해 b값 업데이트

  if i%100 == 0:
    print('epoch=%.f, 기울기a1=%.04f, 기울기a2=%.04f, 절편=%.04f' % (i, a1, a2, b))
```
