```
import numpy as np
import matplotlib.pyplot as plt

# x : 공부한 시간, y = 받은 성적
data=[[2, 81],[4, 93],[6, 91],[8, 97]]
x=[i[0] for i in data]
y=[i[1] for i in data]

# 그래프로 그리기
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.show()
```

```
# 리스트 데이터를 np.array 자료형으로 변경
x_data=np.array(x)
y_data=np.array(y)

# 기울기와 y 절편 초기화
a=0
b=0

# 학습률
lr=0.03

# 반복횟수
epochs=2001

# 경사하강법
for i in range(epochs):   # 에폭의 수 만큼 반복
  y_hat=a*x_data+b   # y를 구하기
  error=y_data-y_hat  # 오차를 구하는 식
  
  a_diff = -(2/len(x_data)) * sum(x_data * error)
  b_diff = -(2/len(x_data)) * sum(error)

  a = a - lr * a_diff # 학습률을 곱해 a값 업데이트
  b = b - lr * b_diff # 학습률을 곱해 b값 업데이트

  if i%100 == 0:
    print('epoch=%.f, 기울기=%.04f, 절편=%.04f' % (i, a, b))
    y_pred=a*x_data+b

    plt.scatter(x,y)
    plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
    plt.show()
```

```
y_pred=a*x_data+b

plt.scatter(x,y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```
